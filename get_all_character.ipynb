```python
import requests
from bs4 import BeautifulSoup

response = requests.get("https://sites.google.com/site/lfostory/index2/character")

soup = BeautifulSoup(response.text, "html.parser")
htmlPrettify = soup.prettify()
```

```python
print(htmlPrettify)
```

```python
charTitle = soup.find(text="點擊角色頭像開始查詢")
```

```python
non_r_char = charTitle.next_element.div
non_r_char
non_r_char_children = non_r_char.findChildren("a", recursive=False)
non_r_char_names = [child["href"].split("/")[-1] for child in non_r_char_children]
print(non_r_char_names)
```

```python
r_char = charTitle.next_element.select("div:nth-of-type(2) > div:nth-of-type(3) ")[0]
r_char
r_char_children = r_char.findChildren("a", recursive=False)
r_char_names = [child["href"].split("/")[-1] for child in r_char_children]
print(r_char_names)
```

```python
all_char_names = non_r_char_names + r_char_names
all_char_names
```
